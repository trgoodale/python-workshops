# Sorting

```python
mtcars.sort_values('cyl')
```

By default, the sort is ascending.  We can also sort descending.

```python
mtcars.sort_values('cyl', ascending=False)
```

We can also set different sort orders for each variable we're sorting by.

```python
mtcars.sort_values(['cyl', 'gear'], ascending=[False, True])
```

# Aggregation

The **.groupby** method allows us to produce summaries based upon subsets of the dataset.

```python
mtcars.groupby('cyl').mean()
```

```python
(mtcars.groupby('cyl'))['mpg'].mean()
```

```python
mtcars.groupby(['cyl', 'am']).mean()
```

## Pivot tables

```python
mtcars.pivot_table('mpg', index='cyl', columns='am')
```

By default it uses **np.mean()** to summarise the data, however we can use any function.

```python
mtcars.pivot_table('mpg', index='cyl', columns='am', aggfunc=np.std)
```

We can use more than one function.

```python
mtcars.pivot_table('mpg', index='cyl', columns='am', aggfunc=[np.mean,np.std])
```

and produce summaries of more than one variable.

```python
mtcars.pivot_table(['mpg','wt'], index='cyl', columns='am', aggfunc=[np.mean,np.std])
```

We don't need to use the same summary function or functions for each variable.

```python
mtcars.pivot_table(['mpg','wt'], index='cyl', columns='am', aggfunc={'mpg': np.mean, 'wt':[np.mean,np.std]})
```

## Using the **.apply** method

```python
mtcars.head()
```

```python
mtcars.iloc[:,0:3]
```

Let's just make a subset of the dataframe to work with.

```python
mycars_n = mtcars.iloc[:,0:3]
```

```python
mycars_n.head()
```

Taking the square root of everything in our subset.

```python
mycars_n.apply(np.sqrt).head()
```

If it is an aggregation function, it will apply it to whole columns.

```python
mycars_n.apply(np.sum)
```

```python
mycars_n.apply(np.mean)
```

We can also ask it to apply it to rows instead.

```python
mycars_n.apply(np.sum, axis=1).head()
```

```python
mycars_n.head()
```

If there isn't a standard function, we can create one and use it.

```python
def myfunc(row):
    return row[0] + row[1]

mycars_n.apply(myfunc, axis=1).head()
```

If it's a short, one-use, function, it's quicker to use a "lambda function."

https://www.w3schools.com/python/python_lambda.asp

```python
mtcars.apply(lambda row: row[2]+row[3]+row[4] , axis=1).head()
```

```python
mtcars.apply(lambda row: (row[2]+row[3]+row[4],5) , axis=1).head()
```
